# Machine Learning With Python: Linear Regression With One Variable

# Sample problem of predicting home price in monroe, NJ(USA)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%pip install seaborn 
import seaborn as sns
from sklearn import linear_model
```

```python
data = pd.read_csv("USA_house.csv")
```

```python
data.head()
```

```python
%matplotlib inline
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(data.area , data.price , color='r' , marker='+')
```

```python
new_data = data.drop(columns=["price"])
```

```python
new_data
```

```python
price = data["price"]
```

```python
price
```

```python
regression = linear_model.LinearRegression()
```

```python
regression.fit(new_data, price)
```

```python
regression.predict([[3300]])
```

```python
regression.coef_
```

```python
regression.intercept_
```

# Y = m * X + b (m is coefficient and b is intercept)

```python
plt.xlabel('area', fontsize =20)
plt.title("USA Homes Price")
plt.ylabel('price', fontsize=20)
plt.scatter(data.area , data.price , color='red' , marker='+')
plt.plot(data.area , regression.predict(data[['area']]), color='green')
plt.show()
```
